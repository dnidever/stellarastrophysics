# Classifying Stellar Spectra

```python
# Imports
import numpy as np
from astropy.table import Table
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
""" Load the spectra"""
temp = np.load('spectra.npz')
wave = temp['wave']
sp3500 = {'wave':wave,'flux':temp['sp3500']}
sp5000 = {'wave':wave,'flux':temp['sp5000']}
sp7000 = {'wave':wave,'flux':temp['sp7000']}
sp9000 = {'wave':wave,'flux':temp['sp9000']}
sp20000 = {'wave':wave,'flux':temp['sp20000']}
obs1 = {'wave':wave,'flux':temp['obs1']}
obs2 = {'wave':wave,'flux':temp['obs2']}
```

# Plot our template spectra

```python
fig = plt.figure(figsize=(12,5))
plt.plot(sp3500['wave'],sp3500['flux'])
plt.xlabel('Wavelength (A)')
plt.ylabel('Flux')
plt.title('Teff = 3500')
```

# Classify the two observed spectra with the templates

Note that the observed spectra have some noise and are on a different scale than the template spectra.  This is similar to what real observed spectra look like.

2) You will need to scale the template flux up/down by multiplying the flux by a constant value, i.e. ``spXXXX['flux']*0.5``
3) The observed spectra will likely fall between the templates.  Figure out between which templates each observed spectrum falls and what is the best match

```python
fig = plt.figure(figsize=(12,5))
# Observed spectrum
plt.plot(obs1['wave'],obs1['flux'],label='Observed 1')
# Template spectrum, multiplied by a scaling factor
plt.plot(sp3500['wave'],sp3500['flux']*0.0014,label='template 3500')
plt.xlabel('Wavelength (A)')
plt.ylabel('Flux')
plt.title('Observed Spectrum 1')
plt.legend()
```
