# Arclamp Spectrum

```python
# Imports
import numpy as np
from astropy.table import Table
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_lines(lines,c='r'):
    """ Overplot lines"""
    for i in range(len(lines)):
        plt.axvline(lines['wave'][i],0,42000,c=c)
```

```python
# Load the spectrum
data = Table.read('lampspec.txt',format='ascii')
```

```python
# Load the linelists
argon = Table.read('argon.dat',format='ascii',names=['wave','name'])
helium = Table.read('helium.dat',format='ascii',names=['wave','name'])
iron = Table.read('iron.dat',format='ascii',names=['wave','name'])
neon = Table.read('neon.dat',format='ascii',names=['wave','name'])
thorium = Table.read('thorium.dat',format='ascii',names=['wave'])
```
