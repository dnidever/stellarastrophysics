```python
#luminosity = np.arange(0.8,1.2,0.01)
#temperature = np.arange(5000.0,6000.0,10)
luminosity = np.arange(0.8,1.2,0.02)
temperature = np.arange(5000.0,6000.0,50)
igoof = np.zeros((len(luminosity),len(temperature)),int)
mass0 = np.zeros((len(luminosity),len(temperature)),float)
lum0 = np.zeros((len(luminosity),len(temperature)),float)
temp0 = np.zeros((len(luminosity),len(temperature)),float)
for i,lum in enumerate(luminosity):
    for j,teff in enumerate(temperature):
        Igoof,ierr,istop,tab = starmodel(1.0,lum,teff,0.70,0.008,verbose=False)
        igoof[i,j] = Igoof
        mass0[i,j] = tab['M'][-1]
        lum0[i,j] = tab['L'][-1]
        temp0[i,j] = tab['T'][-1]
        print(lum,teff,Igoof,mass0[i,j],lum0[i,j],temp0[i,j])
```

```python
print(len(luminosity))
print(len(temperature))
print(igoof.shape)
```

```python
plt.figure(figsize=(10,6))
plt.imshow(mass0,extent=[temperature[0],temperature[-1],luminosity[0],luminosity[-1]],
           origin='lower',aspect='auto')
plt.colorbar()
```

```python
plt.figure(figsize=(10,6))
plt.imshow(lum0,extent=[temperature[0],temperature[-1],luminosity[0],luminosity[-1]],
           origin='lower',aspect='auto')
plt.colorbar()
```

```python
plt.figure(figsize=(10,6))
plt.imshow(temp0,extent=[temperature[0],temperature[-1],luminosity[0],luminosity[-1]],
           origin='lower',aspect='auto')
plt.colorbar()
```
