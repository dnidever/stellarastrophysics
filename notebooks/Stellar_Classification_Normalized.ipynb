# Classifying Normalized Stellar Spectra

```python
# Imports
import numpy as np
from astropy.table import Table
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
""" Load the spectra"""
temp = np.load('spectra_normalized.npz')
wave = temp['wave']
sp3500 = {'wave':wave,'flux':temp['sp3500']}
sp5000 = {'wave':wave,'flux':temp['sp5000']}
sp7000 = {'wave':wave,'flux':temp['sp7000']}
sp9000 = {'wave':wave,'flux':temp['sp9000']}
sp20000 = {'wave':wave,'flux':temp['sp20000']}
obs1 = {'wave':wave,'flux':temp['obs1']}
obs2 = {'wave':wave,'flux':temp['obs2']}
obs3 = {'wave':wave,'flux':temp['obs3']}
```

# Plot our template spectra

```python
fig = plt.figure(figsize=(12,5))
plt.plot(sp3500['wave'],sp3500['flux'])
plt.xlabel('Wavelength (A)')
plt.ylabel('Flux')
plt.title('Teff = 3500')
```

```python
fig = plt.figure(figsize=(12,5))
plt.plot(sp3500['wave'],sp5000['flux'])
plt.xlabel('Wavelength (A)')
plt.ylabel('Flux')
plt.title('Teff = 5000')
```

```python
fig = plt.figure(figsize=(12,5))
plt.plot(sp3500['wave'],sp7000['flux'])
plt.xlabel('Wavelength (A)')
plt.ylabel('Flux')
plt.title('Teff = 7000')
```

```python
fig = plt.figure(figsize=(12,5))
plt.plot(sp3500['wave'],sp9000['flux'])
plt.xlabel('Wavelength (A)')
plt.ylabel('Flux')
plt.title('Teff = 9000')
```

```python
fig = plt.figure(figsize=(12,5))
plt.plot(sp3500['wave'],sp20000['flux'])
plt.xlabel('Wavelength (A)')
plt.ylabel('Flux')
plt.title('Teff = 20000')
```

# Plot our observed spectra

```python
fig = plt.figure(figsize=(12,5))
plt.plot(obs1['wave'],obs1['flux'])
plt.xlabel('Wavelength (A)')
plt.ylabel('Flux')
plt.title('Observed Spectrum 1')
```

```python
fig = plt.figure(figsize=(12,5))
plt.plot(obs2['wave'],obs2['flux'])
plt.xlabel('Wavelength (A)')
plt.ylabel('Flux')
plt.title('Observed Spectrum 2')
```
