# Isochrone Fitting

```python
# Imports
import numpy as np
import dill as pickle
from astropy.table import Table
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
""" Load the data """
with open('isochrone_fitting.pkl', 'rb') as inp:
    isochrone = pickle.load(inp)
    obs1 = pickle.load(inp)
    obs2 = pickle.load(inp)
```

## Isochrone Object
`isochrone` is an object that you can call as a function to return the isochrone for a desired set of age (in Gyr) and metallicity.  This does not perform interpolation.  You need to pick an age and metallicity from the above list.  Here's an example:

```python
# Isochrone information for age=2.5 Gyr and metallicity = -1.5
iso = isochrone(2.5,-1.5)
```

The returned isochrone table contains absolute magnitudes in various photometric bands (e.g., bmag, vmag) that we will compare to the observed apparent magnitudes in the color-magnitude diagram (CMD).

# Plot some of the isochrones

```python
fig = plt.figure(figsize=(7,8))
iso1 = isochrone(0.5,-1.5)
plt.scatter(iso1['bmag']-iso1['vmag'],iso1['vmag'],s=5)
plt.ylim(10,-6)
plt.xlabel('B-V')
plt.ylabel('Absolute V Magnitude')
plt.title('obs1 CMD')
```

```python
fig = plt.figure(figsize=(7,8))
iso2 = isochrone(5.0,-0.5)
plt.scatter(iso2['bmag']-iso2['vmag'],iso2['vmag'],s=5)
plt.ylim(10,-4)
plt.xlabel('B-V')
plt.ylabel('Absolute V Magnitude')
plt.title('obs1 CMD')
```

# Plot our observed CMDs

```python
fig = plt.figure(figsize=(7,8))
plt.scatter(obs1['bmag']-obs1['vmag'],obs1['vmag'],s=5)
plt.ylim(23,10)
plt.xlabel('B-V')
plt.ylabel('V')
plt.title('obs1 CMD')
```

```python
fig = plt.figure(figsize=(7,8))
plt.scatter(obs2['bmag']-obs2['vmag'],obs2['vmag'],s=5)
plt.xlim(-0.1,2)
plt.ylim(25,16)
plt.xlabel('B-V')
plt.ylabel('V')
plt.title('obs1 CMD')
```

# Estimate the star cluster age, metallicity and distance
Overplot isochrones and shift them in magnitude until you find the best match.

```python
fig = plt.figure(figsize=(7,8))
plt.scatter(obs1['bmag']-obs1['vmag'],obs1['vmag'],s=5)
iso1 = isochrone(0.5,-1.5)
plt.scatter(iso1['bmag']-iso1['vmag'],iso1['vmag']+10,s=5,c='r')
plt.ylim(23,5)
plt.xlabel('B-V')
plt.ylabel('V')
plt.title('obs1 CMD')
```
